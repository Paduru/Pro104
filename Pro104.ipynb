```python
import pandas as pd
info = pd.read_csv("SOCR-HeightWeight.csv")
```

```python
weight = info["Weight(Pounds)"].tolist()
print(weight)
```

```python
def weight_mean(wei):
  weight_sum = 0
  for i in wei:
    weight_sum += i

  mean = weight_sum/len(wei)
  print("Mean is " + str(mean) + "lbs")


print(weight)
weight_mean(weight)
```

```python
import csv

weight.sort()
print(weight)
```

```python
def weight_medium(wei):
  n = len(wei)

  if n%2 == 0:
    median1 = float(wei[n//2])
    median2 = float(wei[n//2-1])
    medium = (median1 + median2)/2
  else:
    medium = wei[n//2]

  print("The medium is: " + str(medium))

weight_medium(weight)
```

```python
from collections import Counter
weightData = Counter(weight)
print(weightData)
```

```python
weightDataList = list(weightData.items())
weightDataMostCommon = weightData.most_common()
highest_occurance = weightDataMostCommon[0]
highest_occurance_times = highest_occurance[1]


modes = []

for i in weightDataList:
  if i[1] == highest_occurance_times:
    modes.append(i[0])

print(modes)
```
